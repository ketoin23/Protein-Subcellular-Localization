```python
import torch
from transformers import AutoModel, AutoTokenizer

model_name = "Rostlab/prot_bert_bfd"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)


def generate_bert_embeddings(sequence):
    tokens = tokenizer(sequence, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**tokens)
    embeddings = outputs.last_hidden_state[:, 0, :]
    return embeddings.numpy()


final_df["BERTEmbeddings"] = final_df["Sequence"].apply(generate_bert_embeddings)
```

```python
import numpy as np

def multiply_matrix_by_transpose(matrix):
    matrix = np.array(matrix)
    result = np.dot(matrix.T, matrix)
    return result

final_df['hmm_embed'] = final_df['hmm_prob'].apply(multiply_matrix_by_transpose)

final_df['pssm_embed'] = final_df['pssm_prob'].apply(multiply_matrix_by_transpose)

final_df.head()
```

```python
import tensorflow as tf

tf.keras.backend.set_floatx('float64')
```

```python
from sklearn.model_selection import train_test_split

def tensor_split(X,y, test_size):
    X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=test_size, random_state=7)
    tensor_list = [tf.convert_to_tensor(np_arr, dtype=tf.float32) for np_arr in X_train]

    X_train = tf.stack(tensor_list)

    tensor_list2 = [tf.convert_to_tensor(np_arr, dtype=tf.float32) for np_arr in X_test]

    X_test = tf.stack(tensor_list2)
    return X_train, X_test, Y_train, Y_test
```

```python
def train_nn(X_train, X_test, Y_train, Y_test, epochs ):
  l1 = len(X_train[0])
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(l1, activation='relu'),
    tf.keras.layers.Dense(l1//10, activation='relu'),
    tf.keras.layers.Dense(l1//50, activation='relu'),
    tf.keras.layers.Dense(l1//100, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid'),
  ])
  model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
  early_stopping = tf.keras.callbacks.EarlyStopping(monitor='accuracy', patience=5)

  model.fit(X_train, Y_train, epochs=epochs,callbacks=[early_stopping])
  test_loss, test_acc = model.evaluate(X_test,  Y_test, verbose=2)

  print('\nTest accuracy:', test_acc)

  return model
```
